Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/content/creditcard.csv')
```

```python
#printing first 5 rows
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
#no. of missing values
data.isnull().sum()
```

```python
data['Class'].value_counts()
```

Preprocessing Data

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing the values for transactions
data.groupby('Class').mean()
```

**USING UNDER-SAMPLING**

```python
legit_sample = legit.sample(n=356)
```

```python
#Concatenating the 2 data frames
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X= new_dataset.drop(columns='Class', axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

SPLITTING THE DATASET

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**MODEL TRAINING USING LOGISITIC REGRESSION**

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#ACCURACY of TRAINING DATA
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy Training Data: ", training_data_accuracy)
```

```python
#ACCURACY OF TEST DATA
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("Accuracy Test Data: ", test_data_accuracy)
```
